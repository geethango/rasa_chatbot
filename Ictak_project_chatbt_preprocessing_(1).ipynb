**Seperating the music instruments courses.**

```python
sub='Musical Instruments'
music_df=data[data['subject']==sub]
```

```python

def preprocess_text(text, words_to_include):
    tokens = word_tokenize(text.lower())
    tokens = [token for token in tokens if token.isalpha()]
    tokens = [token for token in tokens if token in words_to_include]
    return tokens


# List of words you want to include in the tokens
words_to_include =['guitar','harmonica', 'piano', 'bass', 'drums','drum'
       'djembe', 'euphonium', 'oud', 'cajon', 'trombone', 'flute','accordion',
       'trumpet', 'clarinet', 'saxophone', 'harp', 'french horn',
       'didgeridoo', 'violin', 'ukulele', 'oboe', 'darbuka', 'brass',
       'banjo']

# Apply the preprocessing function to 'course_title' column and store the result in 'keywords' column
music_df['keywords'] = music_df['course_title'].apply(preprocess_text, words_to_include=words_to_include)

# Flatten the list of tokens
all_tokens = [token for tokens in music_df['keywords'] for token in tokens]

# Create a Pandas Series to use the value_counts() method
word_freq_series = pd.Series(all_tokens)

# Get the most repeated words
most_repeated_words = word_freq_series.value_counts()
most_repeated_words = word_freq_series.value_counts()
most_repeated_words_filtered = most_repeated_words[most_repeated_words > 1]
most_repeated_words_filtered

```

```python
keywords = most_repeated_words_filtered.index.tolist()
```

```python
def find_and_add_keywords(text):
    found_keywords = []
    for keyword in keywords:
        if keyword in text:
            found_keywords.append(keyword)
    return ", ".join(found_keywords)
```

```python
music_df['Topics'] = music_df['keywords'].apply(find_and_add_keywords)
```

```python
music_df['Topics'].unique()
```
